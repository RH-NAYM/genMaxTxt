```python
excel_path = "wav_text_renamed.xlsx"
wav_dir = "wavs"
audio_col = "audio file"

df = pd.read_excel(excel_path)

# Clean Excel filenames
excel_files = (
    df[audio_col]
    .dropna()
    .astype(str)
    .str.strip()
    .tolist()
)

excel_files_set = set(excel_files)

# Read actual audio filenames
audios = os.listdir(wav_dir)

not_found = []

for audio in audios:
    audio = audio.strip()

    if audio not in excel_files_set:
        not_found.append(audio)
        print(audio)

print(f"\nTotal not found: {len(not_found)}")
```

```python
newPath = "newwavs"

if not os.path.exists(newPath):
    os.makedirs(newPath)

    for i in os.listdir(wav_dir):
        if i not in not_found:
            # move audio to new folder
            os.rename(os.path.join(wav_dir, i), os.path.join(newPath, i))
```
